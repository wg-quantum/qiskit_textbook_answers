アダマールゲートが次の2つの形で書けることを示す。

\begin{align*}
 H = \frac{X+Z}{\sqrt2} ≡ \exp(i\frac{\pi}{2}\frac{X+Z}{\sqrt2})
\end{align*}

ここで、≡はグローバル位相の違いを除いて等しいということを示すために使用されている。 したがって、得られたゲートは物理的に等価である。  
ヒント: 固有値が $±1$ の任意の行列Mが、$M^2=I$ を満たしていることと、$e^{i\frac{\pi}{2}M}≡M$ が成り立つことを証明するのが簡単かもしれない。 

a)  
最初の等式 $H = \frac{X+Z}{\sqrt2}$ は  
各ゲートの定義
\begin{align*}
 H=\frac{1}{\sqrt2}\begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix},\quad
 X=\begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix},\quad
 Z=\begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}
\end{align*}
より明らかである。
\begin{align*}
H &= \frac{X+Z}{\sqrt2} \\
  &= \frac{1}{\sqrt2}(\begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} + \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}) \\
  &= \frac{1}{\sqrt2}\begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} \\
\end{align*}

b)  
設問のヒントに従って後半の等式を証明する。 

固有値が $±1$ の任意の行列Mが $M^2=I$ を満たすことを示す。  
固有値+1の固有ベクトル$\ket{q_1}$と固有値-1の固有ベクトル$\ket{q_2}$から成る任意のベクトル$\ket{Q}$に対して $M^2$ を作用させる。  

\begin{align*}
M^2\ket{Q} &= M^2(a\ket{q_1} + b\ket{q_2}) \\
&= M(a\ket{q_1} - b\ket{q_2})  \quad...M\ket{q_1}=\ket{q_1}、M\ket{q_2}=-\ket{q_2} \\
&= a\ket{q_1} + b\ket{q_2} \\
&= \ket{Q} \\
\end{align*}

したがって、任意の行列Mは $M^2=I$ を満たす。


オイラーの公式を拡張させた関数の定義 
\begin{align*}
e^Z = \sum_{n=0}^\infty{\frac{Z^n}{n!}},\quad
\sin(t) = \sum_{n=0}^\infty{\frac{(-1)^nt^{2n+1}}{(2n+1)!}},\quad
\cos(t) = \sum_{n=0}^\infty{\frac{(-1)^nt^{2n}}{(2n)!}} \\
\end{align*}
および$M^2=I$より
\begin{align*}
e^{i\frac{\pi}{2}M}
 &= \sum_{n=0}^\infty{\frac{(i\frac{\pi}{2}M)^n}{n!}} &&...e^Zの定義を利用する。\\
 &= \sum_{n=0}^\infty{\frac{(i\frac{\pi}{2}M)^{2n+1}}{(2n+1)!}} + \sum_{n=0}^\infty{\frac{(i\frac{\pi}{2}M)^{2n}}{(2n)!}} 
 &&...2n+1の場合にM^{2n+1}=M、2nの場合にM^{2n}=Iになる。\\
 &= i\sin(\frac{\pi}{2})M + \cos(\frac{\pi}{2})I &&...sin(t)、cons(t)の定義を利用する。\\
 &= iM
\end{align*}
となる。  
Hは固有値$±1$の行列なので、$H ≡ \exp(i\frac{\pi}{2}H)$が成り立つ。

a)とb)より、アダマールゲートは次の2つの形で書くことができる。

\begin{align*}
 H = \frac{X+Z}{\sqrt2} ≡ \exp(i\frac{\pi}{2}\frac{X+Z}{\sqrt2})
\end{align*}


2

アダマールゲートは次のように rx と rz の操作から構成することが出来る。  

\begin{align*}
 R_x(\theta) = e^{i\frac{\theta}{2}X}, \quad
 R_z(\theta) = e^{i\frac{\theta}{2}Z}, \\
 H ≡ \lim_{n \to \infty}(R_x(\frac{\theta}{n}) R_z(\frac{\theta}{n}))^n. \\
\end{align*}

この等式は適切に選んだθについて成り立つ。  
・θの正しい値を決定する。  

・(正しいθを用いているときに)nが増えるにつれてエラーが二次的に減少していくことを示す。  

トロッター展開を利用して式 $H ≡ \lim_{n \to \infty}(R_x(\frac{\theta}{n}) R_z(\frac{\theta}{n}))^n$ を展開する。  

\begin{align*}
  H &≡ \lim_{n \to \infty}(R_x(\frac{\theta}{n}) R_z(\frac{\theta}{n}))^n \\
  &= \lim_{n \to \infty}(\exp(i\frac{\theta}{2}\frac{X}{n}) \exp(i\frac{\theta}{2}\frac{Z}{n})))^n  \qquad...R_x(\theta) = e^{i\frac{\theta}{2}X}, R_z(\theta) = e^{i\frac{\theta}{2}Z} \\
  &= \lim_{n \to \infty}(\exp(i\frac{\theta}{2}(X+Z)+O(n^{-1}))) \\
  &= \exp(i\frac{\theta}{2}(X+Z)) \\
\end{align*}

問題1より、$H ≡ \exp(i\frac{\pi}{2}\frac{X+Z}{\sqrt2})$ なので、θの値は以下に決定する。  
\begin{align*}
 \theta = \frac{\pi}{\sqrt{2}} \\
\end{align*}

```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
%config InlineBackend.figure_format = 'svg' # 画像をいい感じに表示する
import numpy as np
```

```python
q = QuantumRegister(1)
c = ClassicalRegister(1)

error = {}
for n in range(1,11):

    # 空の量子回路を作る
    qc = QuantumCircuit(q,c)
    
    # 近似アダマールを実装する
    theta = np.pi/np.sqrt(2) # 決定したθの正しい値を使う。
    for j in range(n):
        qc.rx(theta/n,q[0])
        qc.rz(theta/n,q[0])
      
    # 上の近似がどのくらい良いか測る必要がある。これを行う簡単な方法がある。
    # Step 1: 上の近似アダマールをキャンセルする実際のアダマールを使う。
    # 良い近似ならば, 量子ビットは0状態になる。悪い近似ならば、何かしらの重ね合わせ状態になっている。
    qc.h(q[0])
    
    # Step 2: 量子回路を実行して、どのくらい測定値 1 が出るのかを調べる。
    # (良い近似ならば)確実に0 が出るはずなので、1 が出る割合は誤差の指標となる。
    qc.measure(q,c)
    shots = 20000
    job = execute(qc, Aer.get_backend('qasm_simulator'),shots=shots)
    try:
        error[n] = (job.result().get_counts()['1']/shots)
    except:
        pass
        
plot_histogram(error)
```

近似の改善版が次の式から得られる。  

\begin{align*}
 H = \lim_{n \to \infty}(R_z(\frac{\theta}{2n}) R_x(\frac{\theta}{n}) R_z(\frac{\theta}{2n}))^n. \\
\end{align*}

これを実装し、エラーのスケーリングを調べる。 

```python
q = QuantumRegister(1)
c = ClassicalRegister(1)

error = {}
for n in range(1,11):

    # 空の量子回路を作る
    qc = QuantumCircuit(q,c)
    
    # 改善版の近似アダマールを実装する
    theta = np.pi/np.sqrt(2) # 決定したθの正しい値を使う。
    for j in range(n):
        qc.rz(theta/(2*n),q[0])
        qc.rx(theta/n,q[0])
        qc.rz(theta/(2*n),q[0])
      
    # 上の近似がどのくらい良いか測る必要がある。これを行う簡単な方法がある。
    # Step 1: 上の近似アダマールをキャンセルする実際のアダマールを使う。
    # 良い近似ならば, 量子ビットは0状態になる。悪い近似ならば、何かしらの重ね合わせ状態になっている。
    qc.h(q[0])
    
    # Step 2: 量子回路を実行して、どのくらい測定値 1 が出るのかを調べる。
    # (良い近似ならば)確実に0 が出るはずなので、1 が出る割合は誤差の指標となる。
    qc.measure(q,c)
    shots = 20000
    job = execute(qc, Aer.get_backend('qasm_simulator'),shots=shots)
    try:
        error[n] = (job.result().get_counts()['1']/shots)
    except:
        pass
        
plot_histogram(error)
```
