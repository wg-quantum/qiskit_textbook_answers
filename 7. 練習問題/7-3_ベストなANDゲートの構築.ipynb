```python
qr = QuantumRegister(5, 'qr')
cr = ClassicalRegister(1, 'cr')
backend = Aer.get_backend('qasm_simulator')
```

```python
def AND (input1,input2, q_1=0,q_2=1,q_out=2):
    # q_1 が入力1の量子ビット
    # q_2 が入力2の量子ビット
    # q_out が出力
    
    qc = QuantumCircuit(qr, cr)
        
    # 量子ビットq_1, q_2を入力ビットとして準備する。
    if input1=='1':
        qc.x( qr[ q_1 ] )    # 入力1=1の場合、q_1にxゲートを適用する。
    if input2=='1':
        qc.x( qr[ q_2 ] )    # 入力2=1の場合、q_2にxゲートを適用する。
    
    
    # cxゲートを利用する。
    qc.cx(qr[ q_1 ],qr[ q_out ]) # 入力1と出力間でcxゲートを適用する。
    qc.cx(qr[ q_2 ],qr[ q_out ] )# 入力2と出力間でcxゲートを適用する。
    
    if input1=='1' or input2=='1':
        qc.x( qr[ q_out ] )  # 入力1=1もしくは入力2=1の場合、q_outにxゲートを適用する。
    
    # 量子ビットq_outを測定する。
    qc.measure(qr[ q_out ],cr[0]) 
    
    # 回路はシミュレーターで実行されますが、Tenerife のノイズと接続性も再現されるように実行される。
    job = execute(qc, backend, shots=10000, noise_model=noise_model,
                                             coupling_map=coupling_map,
                                             basis_gates=noise_model.basis_gates)
    output = job.result().get_counts()
    
    return output
```

```python
worst = 1
for input1 in ['0','1']:
    for input2 in ['0','1']:
        print('\nProbability of correct answer for inputs',input1,input2)
        prob = AND(input1,input2, q_1=0,q_2=1,q_out=2)[str(int( input1=='1' and input2=='1' ))]/10000
        print( prob )
        worst = min(worst,prob)
print('\nThe lowest of these probabilities was',worst)
```
