# 1-3_量子ビット状態を表現する

1. $|0\rangle$を測定する確率が$1/3$になる状態ベクトルを作成してください。

テキストより、状態$|\psi \rangle$を状態$|x\rangle$で測定する確率は、
$$p(|x\rangle) = | \langle x | \psi \rangle|^2$$
である。したがって、
\begin{align*}
p(|0\rangle) = | \langle 0 | \psi \rangle|^2 = \frac13
\end{align*}
を満たす $\psi$ を求めればよい。$\psi = \psi_1 |0\rangle + \psi_2 |1\rangle$ として計算すると、
\begin{align*}
| \langle 0 | \psi \rangle|^2 &= | \psi_1\langle 0 |0\rangle +  \psi_2\langle 0 |1\rangle|^2\\
&=| \psi_1 |^2\\
\end{align*}
となる。よって、$| \psi_1 |^2 = \frac13$ を満たす $\psi$ が解である。  
ここでは例として、$\psi = \sqrt{\frac13}|0\rangle +  \sqrt{\frac23}|1\rangle$ を解とする。  
実際に測定すると下記のように、$|0\rangle$を測定する確率が$1/3$であることが分かる。

```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram, plot_bloch_vector
from math import sqrt, pi
backend = Aer.get_backend('statevector_simulator')
```

```python
# 初期状態を定義。
initial_state = [1/sqrt(3), sqrt(2/3)]
qc = QuantumCircuit(1)
# `initial_state`状態の0番目の量子ビットを初期化。
qc.initialize(initial_state, 0) 
state = execute(qc,backend).result().get_statevector()
results = execute(qc,backend).result().get_counts()
plot_histogram(results)
```

2. 同じ測定確率を与える別の状態ベクトルを作成してください。

1.と同様に$| \psi_1 |^2 = \frac13$ を満たす $\psi$ が解である。  
1.と異なる状態ベクトルであれば良いため、ここでは例として、$\psi = \sqrt{\frac13}i|0\rangle +  \sqrt{\frac23}|1\rangle$ を解とする。  
実際に測定すると下記のように、$|0\rangle$を測定する確率が$1/3$であることが分かる。

```python
# 初期状態を定義。
initial_state = [1.j/sqrt(3), sqrt(2/3)]
qc = QuantumCircuit(1)
# `initial_state`状態の0番目の量子ビットを初期化。
qc.initialize(initial_state, 0) 
state = execute(qc,backend).result().get_statevector()
results = execute(qc,backend).result().get_counts()
plot_histogram(results)
```

3. これら2つの状態の$|1\rangle$を測定する確率が$2/3$であることを確認します。

1.の$\psi$について計算する。
\begin{align*}
p(|1\rangle) = | \langle 1 | \psi \rangle|^2 &= | \sqrt{\frac13}\langle 1 |0\rangle +  \sqrt{\frac23}\langle 1 |1\rangle|^2\\
&=| \sqrt{\frac23} |^2\\
&= \frac23
\end{align*}
計算結果から、$|1\rangle$を測定する確率が$2/3$であることが分かる。

#### 練習問題

`plot_bloch_sphere_spherical()`を使って以下の状態の量子ビットをプロットしてください。
1. $|0\rangle$

テキストより、量子ビットを下記で表し、$\theta, \phi$の値を求めれば良い。
$$
|q\rangle = \cos{\tfrac{\theta}{2}}|0\rangle + e^{i\phi}\sin{\tfrac{\theta}{2}}|1\rangle
$$

\begin{align*}
|0\rangle &= 1 |0\rangle + 0 |1\rangle\\
&= \cos{0}|0\rangle + e^{i\phi}\sin{0}|1\rangle
\end{align*}
と書けることから、$\theta = 0, \phi \in \R$(任意) が分かる。ブロッホ球にプロットすると下記となる。

```python
from qiskit_textbook.widgets import plot_bloch_vector_spherical
coords = [0,0,1] # [Theta, Phi, Radius] Phiは任意のため、ここでは0としている。
plot_bloch_vector_spherical(coords)
```

2. $|1\rangle$

\begin{align*}
|1\rangle &= 0|0\rangle + 1|1\rangle\\
&= \cos{\frac\pi2}|0\rangle + e^{i\cdot0}\sin{\frac\pi2}|1\rangle
\end{align*}
と書けることから、$\theta = \pi,\ \phi=0 $が分かる。ブロッホ球にプロットすると下記となる。

```python
coords = [pi,0,1] # [Theta, Phi, Radius]
plot_bloch_vector_spherical(coords)
```
